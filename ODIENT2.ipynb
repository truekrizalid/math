```python
import numpy as np
import sympy as sm
from scipy import integrate
import matplotlib.pyplot as plt
% matplotlib inline

fig = plt.figure(figsize = (15,5))
fig.subplots_adjust(wspace = 0.5 ,hspace = 0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

def Sys(X,t=0):
    return np.array([X[1],4*X[0]-X[0]**3])


t = np.linspace(0,100,1000)

# initial values:
iv = np.linspace(-1,1,11)

print(iv)
ax1.plot(0,'b-',label="y1")
ax1.plot(0,'r-',label = 'y2')

for n,i in enumerate(iv):
    Sys0 = np.array([iv[n],iv[-n+1]])
    print(n,i,Sys0)
    X,infodict = integrate.odeint(Sys,Sys0,t,full_output = True)
    x,y = X.T
    # plot
    ax1.plot(x,'b-')
    ax1.plot(y,'r-')
    ax2.plot(x,y,'k')
    
    ax1.set_title("Dynamics in time")
    ax1.set_xlabel("time")
    ax1.grid()
    ax1.legend(loc = "best")

    ax2.set_xlabel("y1")
    ax2.set_ylabel("y2")
    ax2.set_title("Phase space")
    ax2.grid()
   
```

```python
def fact(n):
    fac= 1
    while n >0:
        fac = fac *n
        n -=1
    return fac
```

```python
def leg_an1(n):
    return fact(2*n)/(2**n)/fact(n)**2

def leg_an2(n):
    s = 0 
    an = 1
    while(s<n):
        an = -an*(n-s)*(n+s+1)/(s+2)/(s+1)
        s+=1

    return an
```

```python
def legendre_1(n,X):
    n = int(n)
    pn = 0
    if n%2 == 0:
        M =int(n/2)
    else:
         M =int((n-1)/2)
    for m in range(M+1):
        pn =(-1)**m *fact(2*n-2*m)/(2**n*fact(m))/fact(n-m)/fact(n-2*m)*X**(n-2*m)+pn    
    return pn
```

```python
x = np.linspace(-1,1,21)
legen_4 = legendre_1(4,x)
legen_3 = legendre_1(3,x)
legen_11 = legendre_1(1,x)
legen_22 = legendre_1(2,x)
plt.plot(x,legen_4,"r")
plt.plot(x,legen_3,"b")
plt.plot(x,legen_11,"g")
plt.plot(x,legen_22,"r+")
plt.plot()
plt.show()
```
